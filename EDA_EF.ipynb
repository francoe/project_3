```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import seaborn as sns
```

```python
plt.style.use('seaborn-colorblind')
sns.set_style('whitegrid')
```

```python
master_df = pd.read_csv('complete_data_test1.csv', index_col = 'Year')
rates_data = pd.read_csv('30 Year Mortgage.csv', index_col = 'DATE', parse_dates = True)
income_df = pd.read_csv('test_income.csv', index_col = 'year', parse_dates = True)
income_df = income_df.rename(columns = {'0' : 'avg_income'})
model_output = pd.read_csv('Model_Output.csv')
```

```python
rates_data.loc['2000':'2018'].plot(linewidth = 4, c = '#336699').legend().remove()
plt.grid(linestyle =':')
plt.title('30 Year Fixed Mortage Rates \n annual changes over time', fontname = 'Arial', fontsize = 14)
plt.ylabel('APR',fontname = 'Arial', fontsize = 15)
plt.xlabel('', fontname = 'Arial', fontsize = 15)
plt.ylim(3,8)
plt.axvspan('2011', '2016', color = 'red', alpha = 0.2)
plt.savefig('charts/30_year_rates.png', orientation = 'landscape', dpi = 1800)
plt.show()
```

```python
plt.scatter(x = master_df.loc['2016']['unemployment_rate'],
            y = master_df.loc['2016']['Avg Income']/1000,
           alpha = 0.5
           )
plt.title('Income vs. Inemployment Rate by Zipcode (2016)', fontname = 'Arial', fontsize = 15)
plt.ylabel("Annual Income ($1000's)",fontname = 'Arial', fontsize = 12)
plt.xlabel('Unemployment Rate', fontname = 'Arial', fontsize = 12)
plt.xlim(0,25)
plt.ylim(10,160)
plt.savefig('charts/income_unemployment_scatter.png', orientation = 'landscape', dpi = 1800)
plt.show()
```

```python
sns.distplot(master_df.loc['2016']['Avg Income']/1000, kde = True, hist = True, color = '#696969')
plt.title('Distribution of Estimated Average Income by Zipcode (2016)', fontname = 'Arial', fontsize = 15)
plt.ylabel("Frequency",fontname = 'Arial', fontsize = 12)
plt.xlabel("Estimated Average Annual Household Income ($000's)", fontname = 'Arial', fontsize = 12)
plt.xlim(0,160)
plt.savefig('charts/income_dist.png', orientation = 'landscape', dpi = 1800)
plt.show()
```

```python
sns.distplot(master_df.loc['2016']['unemployment_rate'], kde = True, hist = True, color = '#696969')
plt.title('Unemployment Rate by Zipcode (2016)', fontname = 'Arial', fontsize = 15)
plt.ylabel("Frequency",fontname = 'Arial', fontsize = 12)
plt.xlabel("Unemployment Rate", fontname = 'Arial', fontsize = 12)
plt.xlim(1,12)
plt.ylim(0,0.4)
plt.savefig('charts/unemployment_dist.png', orientation = 'landscape', dpi = 1800)
plt.show()
```

```python
sns.distplot(master_df.loc['2016']['MedianListPrice']/1000, kde = False, hist = True, color = '#696969', bins = 150)
plt.title('Distribution of Median Listing Price by Zipcode (2016)', fontname = 'Arial', fontsize = 15)
plt.ylabel("Frequency",fontname = 'Arial', fontsize = 12)
plt.xlabel("Listing Price ($000's)", fontname = 'Arial', fontsize = 12)
plt.xlim(0,1000)
plt.ylim(0,2000)
plt.savefig('charts/listing_dist.png', orientation = 'landscape', dpi = 1800)
plt.show()
```

```python
sns.distplot(master_df.loc['2016']['MedianListPrice']/1000, kde = True, hist = False, color = 'r', label = '2016')
sns.distplot(master_df.loc['2015']['MedianListPrice']/1000, kde = True, hist = False, color = 'g', label = '2015')
sns.distplot(master_df.loc['2014']['MedianListPrice']/1000, kde = True, hist = False, color = 'b', label = '2014')
plt.title('Distribution of Median Home value/square foot by Zipcode', fontname = 'Arial', fontsize = 15)
plt.ylabel("Frequency",fontname = 'Arial', fontsize = 12)
plt.xlabel("Listing Price ($000's)", fontname = 'Arial', fontsize = 12)
plt.xlim(0,1000)
plt.savefig('charts/income_dist_by_year.png', orientation = 'landscape', dpi = 1800)
plt.show()
```

```python
plt.scatter(x = master_df.loc['2016']['Avg Income']/1000,
            y = master_df.loc['2016']['MedianListPrice']/1000,
           alpha = 0.5
           )
plt.title('Income vs.Median Listing Price by Zipcode (2016)', fontname = 'Arial', fontsize = 15)
plt.ylabel("Median Listing Price ($000's)",fontname = 'Arial', fontsize = 12)
plt.xlabel("Estimated Average Income($000's)", fontname = 'Arial', fontsize = 12)
plt.savefig('charts/income_price_scatter.png', orientation = 'landscape', dpi = 1800)
plt.xlim(20,160)
plt.show()
```

```python
income_ratio = master_df.loc['2016']['MedianListPrice']/master_df.loc['2016']['Avg Income']

sns.distplot(income_ratio, kde = True, hist = True, color = '#696969')
plt.title('Median Listing price as a multiple of Annual Estimated Income (2016)', fontname = 'Arial', fontsize = 15)
plt.ylabel("Frequency",fontname = 'Arial', fontsize = 12)
plt.xlabel("Listing Price / Annual Income", fontname = 'Arial', fontsize = 12)
plt.xlim(1,12)
plt.ylim(0,0.35)
plt.savefig('charts/income_ratio_dist.png', orientation = 'landscape', dpi = 1800)
plt.show()
```

```python
error = model_output['Prediction']-model_output['Actual']
error = error/1000
_ = model_output['Actual']/1000
pct_error = error/_
```

```python
sns.distplot(error, kde = False, hist = True, color = '#696969')


plt.xlim(-200,200)
```

```python
plt.scatter(x = model_output['Actual']/1000,
            y = error,
           alpha = 0.5
           )
plt.title('Prediction Error vs. Actual Listing Price ($000s)', fontname = 'Arial', fontsize = 15)
plt.ylabel("Error",fontname = 'Arial', fontsize = 12)
plt.xlabel("Listing Price ($000s)", fontname = 'Arial', fontsize = 12)
plt.xlim(0,4000)
plt.ylim(-2000,1000)
plt.savefig('error_v_actual.png', orientation = 'landscape', dpi = 1800)
plt.show()
```

```python
plt.scatter(x = model_output['Actual']/1000,
            y = pct_error,
           alpha = 0.5
           )
plt.title('% Prediction Error vs. Actual Listing Price ($000s)', fontname = 'Arial', fontsize = 15)
plt.ylabel("Error",fontname = 'Arial', fontsize = 12)
plt.xlabel("Listing Price ($000s)", fontname = 'Arial', fontsize = 12)
plt.xlim(0,4000)
plt.savefig('pct_error_v_actual.png', orientation = 'landscape', dpi = 1800)
plt.show()
```
