```python
import pandas as pd
```

Unemployment Data by county (https://www.bls.gov/lau/#tables)

```python
list_ = []
for x in range(2010,2019):
    #print(str(x)[-2:])
    df = pd.read_fwf(f'https://www.bls.gov/lau/laucnty{str(x)[-2:]}.txt', skiprows = 5)
    list_.append(df)
all_data = pd.concat(list_)
all_data = all_data.dropna()
all_data = all_data.rename(columns = {
    'Unnamed: 3' : 'County',
    'Unnamed: 4' : 'Year',
    'Unnamed: 8' : 'Unemployment_Rate'
})
```

```python
all_data.Year = all_data.Year.astype(int)
all_data[['Year','County','Unemployment_Rate']]
```

Income Data by County (https://www.irs.gov/statistics/soi-tax-stats-individual-income-tax-statistics-zip-code-data-soi)

```python
df = pd.read_csv('https://www.irs.gov/pub/irs-soi/16zpallagi.csv')[['zipcode', 'agi_stub', 'N1']]
```

```python
df['income'] = 0
df['income_share'] = 0
```

```python
for index, row in df.iterrows():
    if row['agi_stub'] == 1:
        row['income'] = 12500
    elif row['agi_stub'] == 2:
        row['income'] = 37500
    elif row['agi_stub'] == 3:
        row['income'] = 62500
    elif row['agi_stub'] == 4:
        row['income'] = 87500
    elif row['agi_stub'] == 5:
        row['income'] = 150000
    elif row['agi_stub'] == 6:
        row['income'] = 200000
```

```python
for index, row in df.iterrows():
    row['income_share'] = row['n1'] / row[]
```

```python
zipcodes = pd.DataFrame(df.zipcode.unique())
zipcodes = zipcodes.rename(columns = {
    0 : 'Zipcode'
})
```
