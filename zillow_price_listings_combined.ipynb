```python
%matplotlib inline

# import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import datetime 
import seaborn as sns
```

```python
# setup file path
zillow_listings ="C:/Users/maque/Dropbox/Rice Data analytics classes/RICEHOU201811DATA2/Project 3/MonthlyListings_AllHomes.csv"
```

```python
# read csv file
zillow_monthly_listings = pd.read_csv(zillow_listings, low_memory=False)
zillow_monthly_listings.head()
```

```python
# setup file path
zillow_price ="C:/Users/maque/Dropbox/Rice Data analytics classes/RICEHOU201811DATA2/Project 3/zillow_price_sqft.csv"
```

```python
# read csv file
zillow_price_sqft = pd.read_csv(zillow_price, low_memory=False)
zillow_price_sqft.head()
```

```python
# time to combine listings and prices
# pd.merge(a, b, on=['A', 'B'])
zillow_price_listings = pd.merge(zillow_price_sqft, zillow_monthly_listings, on=['Date','RegionID'], how ='left')
zillow_price_listings.head()
```

```python
# drop extra columns
zillow_df = zillow_price_listings.drop(['SizeRank_y', 'RegionName_y','StateName','SizeRank_y'], axis=1)
zillow_df.head()
```

```python
# rename columns
zillow_df = zillow_df.rename(columns={'RegionID': 'ZipCode', 
                                      'RegionName_x': 'ZipCodeName', 
                                      'SizeRank_x': 'SizeRank'})
zillow_df.head()
```

```python
# reorder column sequence
zillow_df = zillow_df[['Date', 
                       'RegionType',
                       'ZipCodeName', 
                       'ZipCode', 
                       'City',
                       'State',
                       'Metro',
                       'CountyName',
                       'SizeRank',
                       'MedianValueSqft',
                       'Monthly Listings']]
zillow_df.head()
```

```python
# rename "Monthly Listings" as "MonthlyListings"
zillow_df = zillow_df.rename(columns={'Monthly Listings': 'MonthlyListings'})
zillow_df.head()
```

```python
zillow_df['Date'].dtype
```

```python
# convert datatype of the date to datetime
zillow_df['Date'] = pd.to_datetime(zillow_df['Date'])
zillow_df.head()
```

```python
# add year columns
year = zillow_df['Date'].dt.year
zillow_df['Year'] = year
```

```python
zillow_df.head()
```

```python
# by year average
years_grouped = zillow_df.groupby(['Year','ZipCode'])
print (years_grouped[['MedianValueSqft','MonthlyListings']].agg(np.mean))
```

```python
price_listings_byYear = (years_grouped[['MedianValueSqft','MonthlyListings']].agg(np.mean))
price_listings_byYear.head()
```

```python
# save as csv file
price_listings_byYear.to_csv('C:/Users/maque/Dropbox/Rice Data analytics classes/RICEHOU201811DATA2/Project 3/price_listings_byYear.csv', encoding='utf-8', index=True)
```
